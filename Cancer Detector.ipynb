```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
dataset = sklearn.datasets.load_breast_cancer()
dataset
```

```python
data_frame = pd.DataFrame(dataset.data, columns = dataset.feature_names)
data_frame
```

```python
data_frame['label'] = dataset.target
data_frame
```

```python
data_frame.shape
```

```python
data_frame.apply(pd.isnull).sum()
```

```python
data_frame.describe()
```

```python
data_frame.label.value_counts()
```

```python
x_data= data_frame.drop(columns='label', axis=1)
y_data= data_frame['label']
```

```python
import matplotlib.pyplot as plt
plt.plot(x_data, y_data)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=2)
```

```python
print(x_train.shape, x_test.shape,x_data.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
plt.plot(x_train, y_train)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_train_predict=model.predict(x_train)
training_accuracy = accuracy_score(y_train_predict, y_train)
training_accuracy
```

```python
y_test_predict=model.predict(x_test)
test_accuracy = accuracy_score(y_test_predict, y_test)
test_accuracy
```

```python
def prediction_fuction(input_data, output):
    input_data_scaled = scaler.transform(input_data)
    prediction = model.predict(input_data_scaled)
    if (prediction == 0):
        print('The tumor is Malignant')
        if(output== 'M') :
            print('Prediction is correct as per original data')
        else:
            print('Prediction is incorrect as per original data')
    else:
        print('The tumor is Benign')
        if(output== 'B') :
            print('Prediction is correct as per original data')
        else:
            print('Prediction is incorrect as per original data')
```
